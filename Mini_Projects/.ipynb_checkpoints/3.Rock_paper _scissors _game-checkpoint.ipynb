Rock paper scissors

You can use different function for 

**Rock wins against scissors; paper wins against rock; and scissors wins against paper.**

```python
import random
```

```python
def user(number,mylist):
    return mylist[number]
    
    
```

```python
def computer(mylist):
    random.shuffle(mylist)
    return mylist[0]
```

```python
print("\n**Welcome to rock paper scissors Game****")

mylist=['rock','paper','scissors']

user_win_count=0
com_win_count=0

while True:
    
    
    try:
        num=int(input("\nUsers choice-->\n Choose any one:\n 1.Rock\n 2.Paper \n 3.Scissors\n 4.Quit\n"))
        
        if num>4:       # for correcting the error
            print(3+'4')
        
        else:
            if num==4:
                print("Thanks for playing!!!!!")
                
                # If this choice get select then it computes all the wins and display it 
                
                print(f"\nuser wins: {user_win_count} vs computer wins: {com_win_count}  ")
                break

            else:
                pass
                
                # rock:- 0 ,paper:- 1, scissors:- 2

                user_check=user(num-1,mylist)  # ask user to enter the choice

                comp_check=computer(mylist)
                play_again=True
                
                print(f"\nuser choice:- {user_check}")
                print(f"\n computer choice:- {comp_check}")
                
                
       
        
    
    except:
        print("\nPlease select the valid option! Play again")
        continue   #until try does become correct it will  run again and  again
        
    
    

   
    
    while play_again:
        
        #Rock wins against scissors; paper wins against rock; and scissors wins against paper.
        
        if user_check=='rock' and comp_check=='scissors':
            print("\n*********************************")
            print("\t \nuser wins")
            user_win_count+=1
            break
        
        elif user_check=='paper' and comp_check=='rock':
            print("\n*********************************")
            print("\t \nuser wins")
            user_win_count+=1
            break
        
        elif user_check=='scissors' and comp_check=='paper':
            print("\n*********************************") 
            print("\t\n user wins")
            user_win_count+=1
            break
        
        elif user_check==comp_check:
            print("\n*********************************")
            print("\nplay again! ,Hard Luck!")
            break
        
        else:
            print("\n*********************************")
            print("\nComputer Wins!")
            com_win_count+=1
            break

    # now check the condition

    #if user_check




        #print(user_check)
       # break
    #print(computer(mylist))
```
